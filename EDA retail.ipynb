```python
correlation = df.corr()
sns.heatmap(correlation,cmap='inferno',annot=True,linewidths=1)
```

Heatmap :
    Profit and sales are highly co-related which means that they have a strong impact on each other.
    Correlation between quantity and profit is weak,profit is not influenced heavily by the quantity of goods.
    Sales of Furniture of Category are high,but profits generated are low compared to the office supplies,Tech.
    Bookcases,tables have higer sales and operating at loss.Limiting the sales would be beneficial.
